```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
reclamations=pd.read_csv("reclamations_V1.csv")
```

```python
grouped = reclamations.groupby('Sexe')['DESCAferPreproc'].count()
```

```python
colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow']
grouped.plot(kind='bar', color=colors)
plt.xlabel('Sexe')
plt.ylabel('Nombre de réclamations')
plt.title('Nombre de réclamations par Sexe')
plt.show()
```

```python
grouped = reclamations.groupby('Type_marche')['DESCAferPreproc'].count()
colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow']
grouped.plot(kind='bar', color=colors)
plt.xlabel('Type_marche')
plt.ylabel('Nombre de réclamations')
plt.title('Nombre de réclamations par Type_marche')
plt.show()
```

```python
grouped = reclamations.groupby('Type_personne')['DESCAferPreproc'].count()
colors = ['red', 'blue','green', 'cyan', 'magenta', 'yellow']
grouped.plot(kind='bar', color=colors)
plt.xlabel('Type_personne')
plt.ylabel('Nombre de réclamations')
plt.title('Nombre de réclamations par Type_personne')
plt.show()
```

```python
grouped = reclamations['DESCAferPreproc'].groupby(reclamations['Secteur_activite']).count()
colors = ['red', 'blue','green', 'cyan', 'magenta', 'yellow']
grouped.plot(kind='bar', color=colors)
plt.xlabel('Secteur_activite')
plt.ylabel('Nombre de réclamations')
plt.title('Nombre de réclamations par Secteur_activite')
plt.show()
```

```python
grouped = reclamations.groupby('Tranche_age')['DESCAferPreproc'].count()
# Créer le diagramme en camembert
plt.pie(grouped, labels = grouped.index, autopct='%1.1f%%')
plt.title('Pourcentage des réclamations par Tranche_age')
plt.show()
```

```python
for age_range in reclamations['Tranche_age'].unique():
    subset = reclamations[reclamations['Tranche_age'] == age_range]
    text = ' '.join(DESCAferPreproc for DESCAferPreproc in subset['DESCAferPreproc'])
    wordcloud = WordCloud().generate(text)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Nuage de mots pour la tranche d\'âge : {age_range}')
    plt.axis('off')
    plt.show()
```
